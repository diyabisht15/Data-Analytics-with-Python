```python
# Import Numpy and check version
import numpy as np
print("NumPy version:", np.__version__)
```

```python
#1) Array Creation & Basics
# 1D Array
arr1 = np.array([1, 2, 3, 4, 5])
print("1D Array:", arr1)

# 2D Array (matrix)
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array:\n", arr2)

# Range of numbers
arr3 = np.arange(0, 11, 2)   # from 0 to 10 step 2
print("Arange:", arr3)

# Evenly spaced numbers
arr4 = np.linspace(0, 1, 5)  # 5 numbers between 0 and 1
print("Linspace:", arr4)

# Random numbers
arr5 = np.random.randn(5)  # 5 random numbers from normal dist
print("Random:", arr5)
```

```python
#2) Array Operations
a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])

print("Addition:", a + b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Exponentiation:", a ** 2)

# Broadcasting (scalar applied to whole array)
print("a + 10:", a + 10)
```

```python
#3) Statistics with NumPy
data = np.random.randint(1, 100, size=20)  # 20 random ints between 1–100
print("Data:", data)

print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Standard Deviation:", np.std(data))
print("Variance:", np.var(data))
print("Min:", np.min(data))
print("Max:", np.max(data))
```

```python
#4) Linear Algebra with NumPy
# Matrix multiplication
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print("Matrix A:\n", A)
print("Matrix B:\n", B)

C = np.dot(A, B)
print("A dot B:\n", C)

# Solve system of equations: 2x + y = 5, x – y = 1
coeff = np.array([[2, 1], [1, -1]])
const = np.array([5, 1])
solution = np.linalg.solve(coeff, const)
print("Solution (x, y):", solution)

# Eigenvalues & Eigenvectors
eigvals, eigvecs = np.linalg.eig(A)
print("Eigenvalues:", eigvals)
print("Eigenvectors:\n", eigvecs)
```

```python
#5) Simple Linear Regression (from scratch)
# True relationship: y = 3x + 7 + noise
np.random.seed(42)  # reproducibility
X = np.random.rand(100, 1) * 10   # 100 values between 0–10
y = 3 * X.flatten() + 7 + np.random.randn(100) * 2
```

```python
#5a) Implement regression formula

# Add bias column (intercept term)
X_b = np.c_[np.ones((100, 1)), X]   # shape (100,2)

# Compute coefficients
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

print("Intercept:", theta_best[0])
print("Slope:", theta_best[1])
```

```python
#5b) Predictions + Error
y_pred = X_b.dot(theta_best)

# Error metrics
mse = np.mean((y - y_pred)**2)
rmse = np.sqrt(mse)

print("MSE:", mse)
print("RMSE:", rmse)
```
