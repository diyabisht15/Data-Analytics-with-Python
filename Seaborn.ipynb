```python
#STEP 1: IMPORT LIBRARIES + LOAD DATA
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Happiness.csv")  # (replace with your file name)

# Clean up column names by removing tab characters
df.columns = df.columns.str.replace('\t', '')

print(df.head())
```

```python
#STEP 2: UNIVARIATE ANALYSIS
# Distribution of Happiness Scores
plt.figure(figsize=(8,5))
sns.histplot(df['Happiness score'], kde=True, bins=20, color="skyblue")
plt.title("Distribution of Happiness Scores")
plt.show()
```

```python
#STEP 3: BIVARIATE ANALYSIS
# Happiness vs GDP per Capita
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="Economy (GDP per Capita)", y="Happiness score")
plt.title("Happiness vs GDP per Capita")
plt.show()
```

```python
#STEP 4: CORRELATION HEATMAP
# Heatmap of numerical columns
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#STEP 5: CATEGORIAL ANALYSIS
# Average Happiness by Region
plt.figure(figsize=(10,6))
sns.barplot(data=df, x="Year", y="Happiness score", hue="Year", palette="viridis", legend=False)
plt.xticks(rotation=45)
plt.title("Average Happiness Score by Year")
plt.show()
```

```python
#STEP 6: MULTIVARIATE ANALYSIS
# Pairwise relationships
sns.pairplot(df[['Happiness score','Economy (GDP per Capita)','Healthy life expectancy','Freedom to make life choices']], diag_kind="kde")
plt.show()
```

```python
#STEP 7: TIME SERIES
# Happiness trend over years (if available)
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x="Year", y="Happiness score")
plt.title("Happiness Trends Over Time")
plt.show()
```
